```python
nums = [1,2,3,4,5,6,7]
print(nums)
```

```python
temp = []
for i in nums:
    temp.append(i*2)
print(temp)
```

```python
newlst = [i for i in nums]#expression, for loop, condition(optional)
print(newlst)
```

```python
newlst = [i*2 for i in nums]#expression, for loop
print(newlst)
```

```python
sqlist = [i*i for i in nums]
print(sqlist)
```

```python
countries = ["india", "uk","france","italy"]
print(countries)
```

```python
upperlist = [j.upper() for j in countries]
print(upperlist)
```

```python
dblist = [i*2 for i in [1,2,3,4,5]]
print(dblist)
```

```python
dblist = [i*2 for i in range(0,10)]
print(dblist)
```

```python
dblist = [i*2 for i in range(0,10) if i<6]
print(dblist)
```

```python
dblist = [i*2 for i in range(0,10) if i>=5 and i<=8]
print(dblist)
```

```python
a = "1a2b3c4gd"
print(a)
```

```python
numbers = [int(i) for i in a if i.isnumeric()]#normal i is will give numbers in string, but typecasting it to int will give numbers

numbers = [i for i in a if i.isnumeric()]#normal i is will give numbers in string, but typecasting it to int will give numbers
print (numbers)
```

```python
alphas = [i for i in a if i.isalpha()]
print(alphas)
```

```python
numbers = [i for i in a if i.isnumeric()]
num2 = int("".join(numbers))#"" is separator, then join numbers, then at last type conversion
print(num2)
```

```python
alphas = [i for i in a if i.isalpha()]
alpha2 = "".join(alphas)
print(alpha2)
```

```python
alphas = [i for i in a if i.isalpha()]
alpha2 = "-".join(alphas)
print(alpha2)
```
