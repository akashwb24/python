```python
import datetime as dt
import dateutil.relativedelta as rel
```

```python
d1 = dt.date(2010,1,1)
d2 = dt.date.today()
rel.relativedelta(d2,d1)#gives difference
```

```python
d1 + rel.relativedelta(years=5)#years has access to all the components for it to add or subtract
```

```python
d1 + rel.relativedelta(months=5)
```

```python
d1 + rel.relativedelta(years=5,months=8,days=10)
```

```python
d1 + rel.relativedelta(year=2022)#here using year instead of years will set the data given
#in case we wnat to add or subtract then we need to use s suffixed names of components
```

```python
dt.date.today() +rel.relativedelta(day=1)
```

```python
dt.date.today() + rel.relativedelta(day=1,months=1,days=-1)
```

```python
dt.date.today() + rel.relativedelta(day=1,days=-1)
```
