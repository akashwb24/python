```python
def add(*args):#here '*args' denotes the number of arguments can be passed, which can be arbitrary argument or it is called variable length argument, multiple arguments can be passed
    print(sum(args))
add(1,2,3,4,5)
```

```python
add(10)
```

```python
def empdet(id, name, city="Delhi", *args):#the variable length argument should be placed at the end of all the arguments, if not then when calling the function we have to explicitly name the parameter and then assign value to it
    print(id)
    print(name)
    print(city)
    print(sum(args))
```

```python
empdet(1,"akash","Kolkata",10,30)
```

```python
def empdetail(id, name, *args, city="Delhi"):
    print(id)
    print(name)
    print(city)
    print(sum(args))
```

```python
empdetail(4,"Rahul",10,20,city="Kolkata")
```

```python
def empdetail2(**kwargs):# '**kwargs' is used to pass multiple arguments as key value pair, notice that the arguments are passed as key value pair
    print(kwargs)
```

```python
empdetail2(id=1,name="John",salary=1000,city="delhi")
```

```python
def empdetails3(**kwargs):
    print(kwargs.values())
```

```python
empdetails3(id=1,name="John",salary=1000,city="delhi")
```

```python
def employeeDetails(empid, empname, *args, city="Mumbai", **kwargs):
    print(empid)
    print(empname)
    print(sum(args))
    print(city)
    print(kwargs)
```

```python
employeeDetails(6,"akash",10,20,city="Kolkata",salary=1000,bonus=2000)
```
