```python
try:
    x=100
    print(x)
    #x/0
    4 + "shhhh"
    y = "hello"
    print(y)
except ZeroDivisionError as zde:
    print(zde)
except TypeError as te:
    print(te)
#only first occurence of error would be shown
```
