```python
lst = list(range(10,60,10))# range can also be used to make an array(list)
print(lst)
```

```python
lst = [10,20,30,40,10,50]
lst.count(10) # provides the number of occurence of any given parameter,
#here 10 comes 2 times in the list
```

```python
lst2 = lst.copy() #no dependabilities are here between lst and lst2
print(lst2)
```

```python
lst2.pop()
print(lst2)
```

```python
lst2.reverse() #reverses the elements in the same list, doesnt give new list
print(lst2)
```

```python
lstt = [95,86,49,68,95,20,45]
lstt.sort()#default sort is in ascending order
print(lstt)#sort the list in the same list and doesnt give new list
lstt.sort(reverse=True)#if we want the list to be sorted in descending order then argument should be passed as "reverse=True"
print(lstt)
```

```python
countries = ['india', 'uk', 'usa', 'russia', 'france']
countries.sort()
print(countries)
countries.sort(reverse=True)
print(countries)
```
