```python
import datetime as dt
```

```python
d1 = dt.datetime(2022,4,12,12,34,18)
print(d1)
```

```python
yr = d1.strftime("%y")#always return string
print(yr)
```

```python
yr = d1.strftime("%Y")
print(yr)
```

```python
mn = d1.strftime("%m")
print(mn)
#small m for month and capital M for minutes
```

```python
dy = d1.strftime("%d")
print(dy)
```

```python
hr = d1.strftime("%H")
print(hr)
```

```python
minu = d1.strftime("%M")
print(minu)
```

```python
secs = d1.strftime("%S")
print(secs)
```

```python
mnname3letter = d1.strftime("%b")
print(mnname3letter)
#%b gives short name of month
```

```python
mnnamefull = d1.strftime("%B")
print(mnnamefull)
```

```python
dynameshort = d1.strftime("%a")
print(dynameshort)
```

```python
dynamefull = d1.strftime("%A")
print(dynamefull)
```

```python
d1.strftime("%d-%m-%Y")
```

```python
d2 = "15-03-2022"
print(d1)
```
