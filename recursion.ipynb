```python
def sumnum(n):
    total =0
    for i in range(1,n+1):
        total = total + i
    return total
```

```python
sumnum(4)
```

```python
def sumnumbersrec(num):
    if num ==1:return 1
    return num +sumnumbersrec(num-1)
```

```python
sumnumbersrec(5)
```

```python
#5*4*3*2*1 = 120
def factorial(n):
    total = 1
    for i in range(1,n+1):
        total = total * i
    return total
```

```python
factorial(5)
```

```python
def factrecur(n):
    if n == 1:return 1
    return n * factrecur(n-1)
```

```python
factrecur(5)
```
