```python
def doublenum(x):
    return x*2
doublenum(7)
```

```python
lst = [1,2,3,4,5,6,7,8,9]
print(lst)
```

```python
map(doublenum,lst)#return a map object by default, cant see output directly
```

```python
list(map(doublenum, lst))
```

```python
mp = map(doublenum, lst)
print(mp)
print(list(mp))
```

```python
doublelist = list(map(lambda x : x *2, lst))
print(doublelist)
```

```python
countries = ['india','japan','russia','america','canada']
print(countries)
```

```python
def converttoucase(str):
    return str.upper()
```

```python
map(converttoucase,countries)
```

```python
ucc = list(map(converttoucase,countries))
print(ucc)
```

```python
cnt2 = list(map(lambda str : str.upper(), countries))
print(cnt2)
```
