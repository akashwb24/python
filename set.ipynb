```python
mset1 = {10,20,50,60}
mset2 = {20, 30, 40}
print(mset1.union(mset2))#union removes duplicates from both sets and then gives the output
```

```python
print(mset1.difference(mset2)) #removes common and provides the left set
```

```python
print(mset2.difference(mset1))# removes the common from the left and gives output
```
