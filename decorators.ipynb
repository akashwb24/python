```python
def converttouppercase(func):
    def inner():
        result = func() # call the original function
        return result.upper() # modify the result
    return inner #return the modified function
```

```python
type(converttouppercase)
```

```python
@converttouppercase # applying decorator
def greet():
    return "Hello World"
greet()
```

```python
@converttouppercase
def greet2():
    return "Hello Everyone"
greet2()
```

```python
def squarenum(func):
    def inner(a):
        i = func(a)
        return i*i
    return inner
```

```python
@squarenum
def square(a):
    return a
square(5)
```

```python
square(10)
```

```python
def sqall(func):
    def inner(a):
        result = func(a)
        return [ i * i for i in result]
    return inner
```

```python
@sqall
def squareall(a):
    return a
squareall([1,2,3,4])
```

```python
def cube(fun):
    def inner(a):
        i = fun(a)
        return i*i*i
    return inner
```

```python
@cube
def cube3(a):
    return a
cube3(5)
```

```python
@cube
@squarenum
#we can apply multiple decorators to a function, the inner one will run first and then its output will be given as the parameter to the outer function, and finnally the output of the outer function will be returned, that is value
#can be interchanged, datatype of both the output of functions should be same
def numbersl(a):
    return a
```
