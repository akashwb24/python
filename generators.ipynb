```python
def counter(n):
    i = 1
    while i<=n:
        yield i#pauses execution and remembers the state
        i += 1
```

```python
x = counter(3)
print(x)#cant see values directly as it gives defaultly generator object
```

```python
print(next(x))
```

```python
print(next(x))#resumes execution where it left off
```

```python
g = counter(4)
for i in g:
    print(i)
```
