```python
for i in range (1,11):
    if i <=5:
        print(i)
    else:
        print(f"the loop is running {i}")
```

```python
for i in range (1,11):
    if i <=5:
        print(i)
    else:
        break #loop will terminate
        print(i)
```

```python
for i in range (1, 11):
    print(i)
    break
```

```python
for i in range (1,11):
    if i == 5:
        continue #current iteration is skipped
    print(i)
```

```python
i = 1
while i<= 10:
    if i<=5:
        print(i)
    else:
        break
    i = i+1
```
